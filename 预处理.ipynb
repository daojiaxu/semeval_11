```python
from bert_serving.client import BertClient

bc = BertClient(ip='222.19.197.230', port=5555, port_out=5556,check_version=False)
```

```python
ve_texts = bc.encode(texts_token_list)
```

```python
ve_texts.shape
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(768,)))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
import numpy as np

labels = np.asarray(texts_token_labels_list)
```

```python
history = model.fit(ve_texts,
                    labels,
                    epochs=20,
                    batch_size=100,
                    validation_split=0.2)
```
